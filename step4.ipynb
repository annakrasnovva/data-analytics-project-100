```python
import pandas as pd
```

```python
regs = pd.read_csv('regs.csv')
regs
```

```python
filtered_regs = regs.drop(['email', 'platform', 'registration_type'], axis=1)
filtered_regs.sort_values(['date'], inplace=True)
filtered_regs['date'] = pd.to_datetime(filtered_regs['date']).dt.date
filtered_regs = filtered_regs.groupby('date').agg('count')
filtered_regs = filtered_regs.reset_index()
filtered_regs.columns = ['date', 'registrations']
filtered_regs
```

```python
visits = pd.read_csv('visits.csv')
visits
```

```python
filtered_visits = visits.drop(['platform', 'user_agent'], axis=1)
filtered_visits.sort_values(['date'], inplace=True)
filtered_visits = filtered_visits.groupby('uuid').agg('max')
filtered_visits['date'] = pd.to_datetime(filtered_visits['date']).dt.date
filtered_visits = filtered_visits.reset_index()
filtered_visits = filtered_visits.groupby(['date']).agg('count')
filtered_visits = filtered_visits.reset_index()
filtered_visits.columns = ['date', 'visits']
filtered_visits
```

```python
ads = pd.read_csv('ads.csv')
ads
```

```python
filtered_ads = ads.drop(['utm_source', 'utm_medium'], axis=1)
filtered_ads.sort_values(['date'], inplace=True)
filtered_ads = filtered_ads[filtered_ads['date'] <= '2023-03-08']
filtered_ads['date'] = pd.to_datetime(filtered_ads['date']).dt.date
filtered_ads
```

```python
result = pd.merge(
    filtered_visits,
    filtered_ads,
    left_on='date',
    right_on='date'
)
result  = pd.merge(
    result,
    filtered_regs,
    left_on='date',
    right_on='date',
    how='outer'
)
result = result.fillna(0)
result['registrations'] = result['registrations'].astype('int')
result = result.iloc[:, [0, 1, 4, 3, 2]]
result['date'] = pd.to_datetime(result['date']).dt.date
result
```

```python
result.to_json('ads.json')
```
