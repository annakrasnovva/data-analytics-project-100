```python
import pandas as pd
import numpy as np
```

```python
visits = pd.read_csv('visits.csv')
visits
```

```python
filtered_visits = visits.drop(['user_agent'], axis=1)
filtered_visits.sort_values(['date', 'uuid'], inplace=True)
filtered_visits
```

```python
filtered_visits = filtered_visits.groupby(['uuid', 'platform']).agg('max')
filtered_visits['date'] = pd.to_datetime(filtered_visits['date']).dt.date
filtered_visits = filtered_visits.reset_index()
filtered_visits
```

```python
filtered_visits = filtered_visits.groupby(['date', 'platform']).agg('count')
filtered_visits.columns = ['visits']
#filtered_visits = filtered_visits.reset_index()
filtered_visits
```

```python
regs = pd.read_csv('regs.csv')
regs
```

```python
filtered_regs = regs.drop(['email', 'registration_type'], axis=1)
filtered_regs.sort_values(['date', 'user_id'], inplace=True)
filtered_regs['date'] = pd.to_datetime(filtered_regs['date']).dt.date

filtered_regs = filtered_regs.groupby(['date', 'platform']).agg('count')
filtered_regs.columns = ['registrations']
filtered_regs
```

```python
result = filtered_visits.join(
    filtered_regs,
    how='outer'
)
result = result.fillna(0)
result['visits'] = result['visits'].astype('int')
result['registrations'] = result['registrations'].astype('int')

result['conversion'] = round (result['registrations'] / result['visits'], 2) * 100
result.replace([np.inf, -np.inf], 0, inplace=True)
result['conversion'] = result['conversion'].astype('int')
result = result.reset_index()
result
```

```python
result.to_json('conversion.json')
```
