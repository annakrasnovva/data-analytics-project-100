```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
conversion = pd.read_json('conversion.json')
conversion.columns = ['date_group', 'platform', 'visits', 'registrations', 'conversion']
conversion
```

```python
ads = pd.read_json('ads.json')
ads.columns = ['date_group', 'visits', 'registrations', 'cost', 'utm_campaign']
ads
```

```python
#1

plt.figure(figsize = (10, 7))
ax = sns.barplot(
    x='date_group',
    y='visits',
    data=ads
)
ax.bar_label(ax.containers[0])
plt.title('Total visits')
plt.grid()
plt.show

plt.savefig('charts/Total_visits.png')
```

```python
#2

plt.figure(figsize=(20, 10))

ax = sns.histplot(
    x='date_group',
    weights='visits',
    hue='platform',
    data=conversion,
    multiple='stack',
    bins=7
)

plt.title('Visits by platform')
plt.xlabel('date_group')
plt.ylabel('visits')
plt.grid(True)
plt.show()

plt.savefig('charts/Visits by platform')
```

```python
#3

plt.figure(figsize = (10, 7))

ax = sns.barplot(
    x = 'date_group',
    y = 'registrations',
    data=ads
)

ax.bar_label(ax.containers[0])
plt.title('Total registrations')
plt.grid()
plt.show
```

```python
#4

fig, ax = plt.subplots(figsize=(20, 10))

ax = sns.histplot(
    x='date_group',
    weights='registrations',
    hue='platform',
    data=conversion,
    multiple='stack',
    bins=7
)

plt.title('Registrations by platform')
plt.xlabel('date_group')
plt.ylabel('registrations')
plt.grid(True)
plt.show()

plt.savefig('charts/Regs by platform')
```

```python
#5

conversion_percent = conversion['conversion'].astype(str) + '%'

plt.figure(figsize=(10, 7))

plt.plot(
    conversion['date_group'],
    conversion_percent,
    marker='o'
)

plt.title('Overall conversion')
plt.xlabel('date')
plt.ylabel('conversion')
plt.grid(True)
plt.show()

```
